```python
import bs4
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
```

```python
r = Request('https://www.investing.com/indices/major-indices',headers={"User-Agent": "Mozilla/5.0"})#fool them to be website.
c = urlopen(r).read()
soup = BeautifulSoup(c, "html.parser")
soup
```

```python
soup.find('div',{'class':'js-ga-on-sort'}).find('tbody').find_all('td',{'class':"bold left plusIconTd noWrap elp"})[1].find('a').text
```

```python
def datadic():
    name = []
    lastprice = []
    z = 169
    for i in range(0,5):
        s = 'pid-' + str(z) + '-last'
        x = soup.find('div',{'class':'js-ga-on-sort'}).find('tbody').find_all('td',{'class':"bold left plusIconTd noWrap elp"})[i].find('a').text
        name.append(x)
        y = soup.find('div',{'class':'js-ga-on-sort'}).find('tbody').find('td',{'class':s}).text
        lastprice.append(y)
    return name,lastprice
```

```python
name , lastprice = datadic()
```
